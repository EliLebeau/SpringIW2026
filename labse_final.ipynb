```python
def lemma_counts(column, stop_word_list):
    lemma_count = {}
    for verse in column:
        lemmas = verse.split(' ')
        for lemma in lemmas:
            # lemma = unidecode.unidecode(lemma)
            
            if lemma in lemma_count:
                lemma_count[lemma] +=1
            else:
                lemma_count[lemma] = 1

    word_occurences_table = pd.DataFrame({'Words': lemma_count.keys(), 'Occurrences': lemma_count.values()})
    word_occurences_table = word_occurences_table[~word_occurences_table['Words'].isin(stop_word_list)]
    word_occurences_table.sort_values('Occurrences', ascending=False, inplace=True)
    word_occurences_table.reset_index(inplace=True)
    word_occurences_table.drop('index', axis = 1, inplace=True)

    return word_occurences_table


def greek_lemma_counts(column, stop_word_list):
    lemma_count = {}
    for verse in column:
        lemmas = verse.split(' ')
        for lemma in lemmas:
            new_lemma = unidecode.unidecode(lemma)
            
            if new_lemma in lemma_count:
                lemma_count[new_lemma][1] +=1
            else:
                lemma_count[new_lemma] = [lemma, 1]


    keys = lemma_count.keys()
    words = []
    counts = []
    for key in keys:
        words.append(lemma_count[key][0])
        counts.append(lemma_count[key][1])
    
    


    word_occurences_table = pd.DataFrame({'Words': keys, 'Occurrences': counts, 'Greek' : words})
    word_occurences_table = word_occurences_table[~word_occurences_table['Words'].isin(stop_word_list)]
    word_occurences_table.sort_values('Occurrences', ascending=False, inplace=True)
    word_occurences_table.reset_index(inplace=True)
    word_occurences_table.drop('index', axis = 1, inplace=True)

    return word_occurences_table

hebrew_stop_words = ['כי', 'כל', 'לא', 'על', 'את', 'לי', 'ולא', 'אשר',
                     'אתה', 'עד', 'עלי', 'לך', 'עם', 'אם', 'אליך', 'להם', 'למען',
                     'מאד', 'ואני', 'ממני', 'אל', 'לו', 'בכל', 'מה', 'מי', 'הוא', 'אני', 'אין', 'ואל', 'בך',
                     'כול', 'היי', 'מן', 'סלה']
greek_stop_words_english = [
    'kaio', 'ego', 'eimi', 'o', 'su', 'autao', 'autos', 'tis',
    'eio', 'tos', 'tos', 'tem', 'oiomai', 'emi', 'iemi', 'to',
    'oukeo', 'me', 'tou', 'te', 'pantao', 'suo', 'eisizo', 'aireo',
    'eao', 'prostemi', 'kathemai', 'poieo', 'emas', 'otheo', 'δίω', 'opero', 'orao'
]
potential_stop_words = [
    'tino', 'oida', 'tio', 'epomai', 'auton', 'oudoo', ''
]
latin_stop_words = ['et', 'in', 'me', 'eius', 'non', 'qui', 'est', 'quoniam', 'tua', 'eorum', 'a', 'super', 'mea', 'te',
                  'sunt', 'ad', 'de', 'quia', 'tuam', 'ut', 'meam', 'eum', 'meus', 'eos', 'sicut', 'mihi', 'omnes', 'cum',
                  'tuum', 'tuo', 'tu', 'meum', 'sum', 'tibi', 'mei', 'autem', 'tuos', 'nos', 'ego', 'tuae', 'es', 'tui',
                  'meae', 'ab', 'ne', 'tuis', 'suum', 'meos', 'usque', 'si', 'meo', 'suo', 'suam', 'quae', 'suis',
                  'omnia', 'meis', 'ex', 'enim', 'pro', 'eis', 'quod', 'quis', 'sua',
                  'is', 'tuus', 'suus', 'omnis', 'noster', 'ipse', 'ille', 'hic', 'quasi', 'totus']
```
