```python
import pandas as pd
```

```python
curr_csv = pd.read_csv('./gallicanLemmas/50gallican.csv')
```

```python
lemmatized_verses = []
for file_num in range(50,2600,50):
    curr_csv = pd.read_csv('./septuagintLemmas/'+str(file_num)+'Septuagint.csv')
    for verse_num in range(1, max(curr_csv['LOCATION'])+1):
        curr_verse = curr_csv[curr_csv['LOCATION'] == verse_num]
        verse_string = ''
        for lemma in curr_verse['TITLE']:
            verse_string += lemma[8:]+' '

        lemmatized_verses.append(verse_string[:-1])
```
